# function arguments and parameters


```python
def indore(a):          # a = 10
    print(a)

indore(10)
```

```python
def indore(a):          # a = 10
    print(a)

indore()
```

```python
def add(v):
    v = v + 1
    print(v)

add(5)
```

```python
def addition(x, y):           # x = 10, y = 20
    print(x + y)

addition(10, 20)
```

```python
addition(5)
```

```python
addition(1,2,3)
```

# return statement


```python
s = "engineering"

print(len(s))
```

```python
def play(x,y):
    if x >= y:
        return x
    else:
        return y

print(play(10,20))
```

```python
z = play(5,3)        # z = 5

print(z)
```

```python
def said(t):
    t = t + 2
    return t
    print("finish")

c = said(4)

print(c)
```

```python
def pre():
    print(1)
    print(2)
    return
    print(3)
    print(4)

pre()
```

# default arguments

```python
def say(msg, time = 1):         # msg = "Hi"
    print(msg * time)
```

```python
say("Hi")
```

```python
say("Hi", 5)
```

```python
def linear_regression(test, train, temperature=5):
    split = test * train
    temperature = temperature + 10
    return [split, temperature]
```

```python
linear_regression(10,20)
```
